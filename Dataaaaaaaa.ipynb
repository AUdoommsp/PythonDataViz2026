## **Code** ##

```
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(4))
```

## **acorn_details** ##

**เปิดดูโครงสร้างของข้อมูลทั้งหมดในไฟล์**

```
import pandas as pd
df = pd.read_csv("acorn_details.csv", encoding="latin1")
df.head()

print("\n--- โครงสร้างของข้อมูล ---")
print(df.info())

print("\n--- หมวดหมู่หลักที่มีในไฟล์ ---")
print(df['MAIN CATEGORIES'].unique())
```

**พฤติกรรมการท่องเที่ยวและการพักผ่อน (LEISURE TIME)เพื่อดูว่ากลุ่มประชากรแต่ละแบบชอบไปเที่ยวที่ไหนกันบ้าง**

```
leisure_df = df[df['MAIN CATEGORIES'] == 'LEISURE TIME']

holiday_df = leisure_df[leisure_df['CATEGORIES'] == 'Holiday Destination/Type']

columns_to_show = ['REFERENCE', 'ACORN-A', 'ACORN-B', 'ACORN-C', 'ACORN-Q']
print(holiday_df[columns_to_show])
```

**แผนภูมิแท่ง (Bar Chart) เพื่อเปรียบเทียบ การไปเที่ยวสถานที่ท่องเที่ยวนั้นๆ ระหว่างกลุ่ม ACORN-A (มักเป็นกลุ่มผู้มีรายได้สูง/ฐานะดี) กับ ACORN-Q (กลุ่มอื่นๆ)**

```
import plotly.graph_objects as go

# สร้างกราฟแท่งแบบกลุ่ม (Grouped Bar Chart)
fig = go.Figure()

# เพิ่มแท่งข้อมูลของ ACORN-A
fig.add_trace(go.Bar(
    x=holiday_df['REFERENCE'],
    y=holiday_df['ACORN-A'],
    name='ACORN-A (กลุ่มฐานะดี)',
    marker_color='royalblue'
))

# เพิ่มแท่งข้อมูลของ ACORN-Q
fig.add_trace(go.Bar(
    x=holiday_df['REFERENCE'],
    y=holiday_df['ACORN-Q'],
    name='ACORN-Q (กลุ่มอื่นๆ)',
    marker_color='lightcoral'
))

# ตั้งค่าความสวยงามและการออกแบบ
fig.update_layout(
    title='เปรียบเทียบดัชนีสถานที่ท่องเที่ยวระหว่างกลุ่ม ACORN-A และ ACORN-Q',
    xaxis_title='สถานที่ท่องเที่ยว / รูปแบบการท่องเที่ยว',
    yaxis_title='ดัชนีพฤติกรรม ',
    barmode='group',
    template='plotly_white'
)

# แสดงกราฟ
fig.show()
```
